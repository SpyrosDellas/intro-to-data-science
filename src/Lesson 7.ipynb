```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.stats
from plotnine import *
```

# Part 1 - Line plots using ggplot

```python
def lineplot(hr_year_csv):
    """
    A csv file will be passed in as an argument which contains two columns:
    - 'HR' (the number of homerun hits) and 
    - 'yearID' (the year in which the homeruns were hit).
    
    We will create a chart with points connected by lines, both colored 'red', showing the number of HR by year.
    """
    df = pd.read_csv(hr_year_csv)
    gg = ggplot(df, aes('yearID', 'HR')) + geom_point(color='red') + geom_line(color='red') + \
    ggtitle('Homerun hits by year') + xlab('Year') + ylab('Homerun hits')
    return gg


filename = "../data/hr_year.csv"
lineplot(filename)
```

# Part 2 - Advanced line plots using ggplot

```python
def lineplot_compare(hr_by_team_year_sf_la_csv):
    """Read a csv file and plot it using pandas and ggplot.
    
    The csv file has three columns: yearID, HR, and teamID. The data in the
    file gives the total number of home runs hit each year by the SF Giants 
    (teamID == 'SFN') and the LA Dodgers (teamID == "LAN"). 
    
    We produce a visualization comparing the total home runs by year of the two teams. 
    
    Note that to differentiate between multiple categories on the 
    same plot in ggplot, we can pass color in with the other arguments
    to aes, rather than in our geometry functions. 
    For example, ggplot(data, aes(xvar, yvar, color=category_var))
    """
    df = pd.read_csv(hr_by_team_year_sf_la_csv)
    gg = ggplot(df, aes('yearID', 'HR', color='teamID')) + geom_point() + geom_line() + \
    ggtitle('Homerun hits by year') + xlab('Year') + ylab('Homerun hits')
    return gg
    
    
filename = "../data/hr_by_team_year_sf_la.csv"
lineplot_compare(filename)
```
